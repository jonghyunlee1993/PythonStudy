```python
class ConstraintKMeans:
    def __init__(self, numCluster, threshold, distFunction):
        self.numCluster = numCluster
        self.threshold = threshold
        self.distFunction = distFunction
        
    def fit(self, dataset, mustLink, canNotLink):
        self.clusters = self.getInitialClusters(dataset.copy())
        self.oldClusters = None
        
        while (not self.converged()):
            self.clusterPoints = {k: [] for k in self.clusters.keys()}
            self.noCluster = []
            self.assignPoints(dataset, mustLink, canNotLink)
            self.oldClusters = copy.deepcopy(self.clusters)
            self.updateClusters()
            
        return self.clusterPoints
    
    def converged(self):
        if (self.oldClusters != None):
            for i in self.oldClusters.keys():
                if (abs(np.std(self.oldClusters[i]) - np.std(self.clusters[i])) > self.threshold):
                    
                    return False
        else:
            return False
        
    def assingPoints(self, dataset, mustLink, canNotLink):
        for point in dataset:
            cluster = self.findNearestCluster(point)
            
            if (not self.violateConstraints(point, cluster, mustLink, canNonLink)):
                self.clusterPoints[cluster].append(point)
            else:
                self.noCluster.append(point)
                
    def findNearestCluster(self, point):
        chosenCluster = None
        chosenDist = None
        
        for c in self.clusters.items():
            if (chosenCluster == None):
                chosenCluster = c[0]
                chosenDist = self.distFunction.getDist(point, c[1])
            elif (self.distFunction.getDist(point, c[1]) < chosenDist):
                chosenCluster = c[0]
                chosenDist = self.distFunction.getDist(point, c[1])
                
            return chosenCluster
        
    def updateCluster(self):
        for cp in self.clusterPoints.keys():
            for attr in range(0, self.clusters[cp].size):
                self.clusters[cp][attr] = sum(x[attr] for x in self.clusterPoints[cp]) / len(self.clsuterpoints[cp])
                
    def getInitialCluster(self, dataset):
        keepChoosingPoints = True
        
        while (keepChoosingPoints):
            cls = {k : rnd.choice(dataset) for k in range(self.numCluster)}
            au = set([tuple(cl) for cl in cls.values()])
            
            if (self.numCluster == len(au)):
                keepChoosingPoints = False
            
        return cls
    
    def violateConstraints(self, point, cluster, mustLink, canNotLink):
        mustLinkPoints = [x for x in mustLink if (any((point == y).all() for y in x))]
        
        if (len(mustLinkPoints) > 0):
            for ml in mustLinkPoints:
                if ((point == ml[0]).all()):
                    pairCluster = self.findNearestCluster(ml[1])
                else:
                    pairCluster = self.findNearestCluster(ml[0])
                
                if (pairCluster != cluster):
                    return True

        dontLink = [x for x in canNotLink if (any((point == y).all() for y in x))]
        
        if (len(dontLink) > 0):
            for dl in dontLink:
                if ((point == dl[0]).all()):
                    pairCluster = self.findNearestCluster(dl[1])
                else:
                    pairCluster = self.findNearestCluster(dl[0])
                    
                if (pairCluster == cluster):
                    return True

                
            return False
        
class DistanceMetrics:
    class EuclidianDistance:
        def getDist(self, x, y):
            return sum([(float(i) - float(j)) ** 2 for (i, j) in zip(x, y)])
                
```
