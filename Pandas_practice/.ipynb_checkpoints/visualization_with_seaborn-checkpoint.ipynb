```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# 히스토그램과 밀집도 그래프
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# remove density graph
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# remove histogram
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# add rug graph
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# count graph
ax = plt.subplots()
ax = sns.countplot('day', data = tips)
ax.set_title('Count of Days')
ax.set_xlabel('Day of the Weeks')
ax.set_ylabel('Frequency')
```

```python
# scatter graph
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips)
ax.set_title('Scatterplot of Total Bill and Tips')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# scatter graph without regressor
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tips')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# jointplot
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint PLot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
# jointplot with hex
joint = sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
joint.set_axis_labels(xlabel = 'Total Bill', ylabel = 'Tip')
joint.fig.suptitle('Joint PLot of Total Bill and Tip', fontsize = 10, y = 1.03)
```

```python
# 2D kernel density plot
ax = plt.subplots()
ax = sns.kdeplot(data = tips['total_bill'],
                data2 = tips['tip'],
                shade = True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# bar plot
ax = plt.subplots()
ax = sns.barplot(x = 'time', y = 'total_bill', data = tips)
ax.set_title('Bar plot of average total for time of day')
```

```python
# box plot
ax = plt.subplots()
ax = sns.boxplot(x = 'time', y = 'total_bill', data = tips)
```

```python
# voline plot
ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', data = tips)
```

```python
fig = sns.pairplot(tips)
```

```python
# 대각선을 기준으로 양쪽에 모아주기
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```

```python
# 다변량 그래프, 바이올린 플랏
ax = plt.subplots()
ax = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips, split = True)
```

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = True)
scatter2 = sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', fit_reg = False)
```

```python
fig = sns.pairplot(tips, hue = 'sex')
```

# 그래프 크기와 모양 조절

```python
scatter = sns.lmplot(x = 'total_bill', y = 'tip', fit_reg = 0, data = tips, hue = 'sex',
                    markers = ['o', 'x'], 
                    scatter_kws = {'s' : tips['size']*30, 
                                  'alpha' : 0.5})
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = 0)
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = 0,
                          col = 'dataset', col_wrap = 2)
anscombe_plot = sns.lmplot(x = 'x', y = 'y', data = anscombe, fit_reg = 0,
                          col = 'dataset', col_wrap = 4)
```

```python
# 여러 그래프 한번에 그리기
facet = sns.FacetGrid(tips, col = 'day', hue = 'sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
```

```python
facet = sns.FacetGrid(tips, col = 'time', row = 'smoker', hue = 'sex')
facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
```

```python
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
#         ax = fig.add_subplot(2, 3, plot_position)
        ax = plt.subplots(figsize = (4, 3))
        violin = sns.violinplot(x = 'time', y = 'total_bill', hue = 'sex', data = tips)
        violin.set_title(style)
        
fig.tight_layout()
```
