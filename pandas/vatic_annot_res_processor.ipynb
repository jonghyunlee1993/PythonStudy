```python
import glob
import pandas as pd

glob.glob("*.txt")
```

```python
col_names = ['a', 'x_str', 'y_str', 'x_end', 'y_end', 'frame', 'b', 'c', 'd', 'id']

fir = pd.read_csv("sess_01_fir.txt", sep=" ", names=col_names)
sec = pd.read_csv("sess_01_sec.txt", sep=" ", names=col_names)
thr = pd.read_csv("sess_01_thr.txt", sep=" ", names=col_names)

data = pd.concat([fir, sec, thr], axis=0)
data.head()
```

```python
proc = data.groupby(['id', 'frame']).mean()[['x_str', 'y_str', 'x_end', 'y_end']].astype(int)
proc
```

```python
proc.loc['player_01']
```

```python
players = ['player_01', 'player_02', 'player_03', 'player_04', 'player_05']

for player in players:
    proc.loc[player].to_csv("sess_01_" + player + ".csv", index=False)
```
