```python
# 판매액 
sales_rate_idx = data['FACT'] == '판매액'

# 판매량
sales_amount_idx = data['FACT'] == '판매량 (UNIT)'

```

```python
# 채널별 
sales_rate = data[sales_rate_idx]['합계'].groupby(data['채널'])
print()
print("="*10)
print("채널별 판매액")
print("="*10)
print(sales_rate.sum())

sales_amount = data[sales_amount_idx]['합계'].groupby(data['채널'])
print()
print("="*10)
print("채널별 판매량")
print("="*10)
print(sales_amount.sum())
```

```python
# 채널별 제조사별
sales_rate = data[sales_rate_idx]['합계'].groupby([data['채널'], data['제조사']])
print()
print("="*10)
print("채널별 판매액")
print("="*10)
print(sales_rate.sum())

sales_amount = data[sales_amount_idx]['합계'].groupby([data['채널'], data['제조사']])
print()
print("="*10)
print("채널별 판매량")
print("="*10)
print(sales_amount.sum())
```
